# **Frequency Modulation (FM)**

If the **carrier signal** is **modulated** in a way that it's **Frequency** varies with the changing **Frequency** of modulating signal.<br>
Where **Amplitude and Phase** remain the **same**.

```python
#2017331047
import numpy as np
import matplotlib.pyplot as plt

mod_f = 15.0
carrier_f = 150.0
modulation_index = 1.5
m_s= 58800.0
t = np.arange(m_s) / m_s
modulator = np.sin(2.0 * np.pi * mod_f * t) * modulation_index
carrier = np.sin(2.0 * np.pi * carrier_f * t)
product = np.zeros_like(modulator)

cnt=0
for t in t:
  product[cnt]=np.sin(2. * np.pi * (carrier_f * t + modulator[cnt]))
  cnt=cnt+1

plt.subplot(3, 1, 1)
plt.title('Frequency Modulation')
plt.plot(modulator)
plt.ylabel('Amplitude')
plt.xlabel('Modulator signal')
plt.subplot(3, 1, 2)
plt.plot(carrier,'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')
plt.subplot(3, 1, 3)
plt.plot(product,'r')
plt.ylabel('Amplitude')
plt.xlabel('Output signal')


plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```

# **Amplitude Modulation (AM)**

If the **carrier signal** is **modulated** in a way that it's **Amplitude** varies with the changing **Amplitude** of modulating signal.<br>
Where **Frequency and Phase** remain the **same**. 

```python
#2017331047
import numpy as np
import matplotlib.pyplot as plt

#Carrier wave c(t)=A_c*cos(2*pi*f_c*t)
#Modulating wave m(t)=A_m*cos(2*pi*f_m*t)
#Modulated wave s(t)=A_c[1+mu*cos(2*pi*f_m*t)]cos(2*pi*f_c*t)

A_c = 150
f_c = 80.0
A_m = 120
f_m = 15.0
modulation_index = 1.5

t = np.linspace(0, 1, 1000)

carrier = A_c*np.cos(2*np.pi*f_c*t)
modulator = A_m*np.cos(2*np.pi*f_m*t)
product = A_c*(1+modulation_index*np.cos(2*np.pi*f_m*t))*np.cos(2*np.pi*f_c*t)

plt.subplot(3,1,1)
plt.title('Amplitude Modulation')
plt.plot(modulator)
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product,'r')
plt.ylabel('Amplitude')
plt.xlabel('AM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```

# **Phase Modulation (PM)**

If the **carrier signal** is **modulated** in a way that it's **Phase** varies with the changing **Phase** of modulating signal.<br>
Where **Frequency and Amplitude** remain the **same**. 

[**found the formula from** : [HERE](https://www.allaboutcircuits.com/textbook/radio-frequency-analysis-design/radio-frequency-modulation/phase-modulation-theory-time-domain-frequency-domain/)]

```python
#2017331047
#phase_modulation
import numpy as np
import matplotlib.pyplot as plt

#Carrier wave c(t)=A_c*sin(2*pi*f_c*t)
#Modulating wave m(t)=A_m*sin(2*pi*f_m*t)
#Phase Modulated wave s(t)=A_c*sin[(2*pi*f_m*t)+modulation_index*cos(2*pi*f_c*t)] 

A_c = 100
f_c = 50.0
A_m = 120
f_m = 1.0
modulation_index = 1.0

t = np.linspace(0, 1, 1000)

carrier = A_c*np.sin(2*np.pi*f_c*t)
modulator = A_m*np.sin(2*np.pi*f_m*t)

p_m= A_c*np.sin(2*np.pi*f_c*t+modulation_index*modulator)

plt.subplot(3,1,1)
plt.title('Phase Modulation')
plt.plot(modulator)
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(p_m, 'r')
plt.ylabel('Amplitude')
plt.xlabel('PM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```
