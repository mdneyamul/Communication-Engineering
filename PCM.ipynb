# **PCM**

Pulse Code Modulation(PCM):
It's a process of representing an analog sgnal to a digital signal.<br>
PCM consists of three(3) main processes<br>
1. Sampling: Conversion of Continnuous time signal into Discrete time signal.<br>
2.Quantization:Conversation of Discrete time Continuous valued signal into Discrete time Discrete valued signal.<br>
3.Coding: Each quantized sample is encoded into a binary sequence

```python
#2017331047
import numpy as np
import matplotlib.pyplot as plt
from termcolor import colored
z=np.arange(0,10,0.5);
p=np.linspace(0,9,10);

plt.figure(figsize=(16,9))

#The Input analog signal
a=1.0 # amplitude of the analog signal
f=5.0 #frequency of the analog signal
lim=1.0 #higest value of t in the range
t=np.linspace(-lim,lim,100)
d=-np.pi/2.0 #phase
x_t=a*np.cos(2*np.pi*f*t+d);
plt.subplot(3,1,1)
plt.plot(t,x_t);
plt.title('Input Signal', color ='r',fontsize=22)
plt.grid()

#1st Process: Sampling
d=-np.pi/2.0 #phase

fs=50  #lets assume sampling rate=150
T=1.0/fs; #sample interval
nT=np.arange(-lim,lim,T); #total number of samples n in [-lim,lim] range
x_nT=a*np.cos(2*np.pi*f*nT+d); # Sampled Signal
x_nT[ x_nT==0 ] = np.nan
plt.subplot(3,1,2)  #plotting of that signal
plt.stem(nT,x_nT)
plt.title('Sampling', color ='r',fontsize=22)
plt.grid()

#2nd Process: quantization
bits=2; #Lets assume total number of bits =2
levels= 2**(bits) #total number of levels = 2^bits
print(levels)
step=(2.0/levels)
xq_n=np.round(x_nT/step)*step
xq_n[ xq_n==0 ] = np.nan
print(xq_n[5]) #quantized signal
quantization_code= []
for i in xq_n:
  if i==-1.00:
    quantization_code.append(0)
  elif i==-0.5:
    quantization_code.append(1)
  elif i==0.5:
    quantization_code.append(2)
  elif i==1:
    quantization_code.append(3)

print(quantization_code)
plt.subplot(3,1,3)
plt.stem(nT,xq_n)
plt.title('Quantization', color ='r',fontsize=22)




  

```

# **Encoding**

```python
#2017331047
#3rd Process: Encoding

coded_str=""
for i in quantization_code:
  if i==0:
    coded_str+="00"
  elif i==1:
    coded_str+="01"
  elif i==2:
    coded_str+="10"
  else:
    coded_str+="11"
print(colored("The Coded Binary String is: ", 'red'))
print(coded_str)
  
```
